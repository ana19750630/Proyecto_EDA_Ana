```python
import pandas as pd
```

```python


ruta_del_archivo = "./data/global_food_wastage_dataset.csv" 

# Cargar el archivo CSV en un DataFrame de Pandas
df = pd.read_csv(ruta_del_archivo)
```

```python
df.head()
```

```python
data_2 = df.groupby(["Year", "Food Category", "Country"]).agg({
    "Total Waste (Tons)": "sum"
}).reset_index()
```

```python
data_2.head(90)
```

```python
data_espana = data_2[data_2['Country'] == "Spain"]
data_espana
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar datos solo para España
data_espana = data_2[data_2['Country'] == "Spain"]

# Configurar el gráfico
plt.figure(figsize=(10, 6))
sns.lineplot(data=data_espana, x="Year", y="Total Waste (Tons)", marker="o", color="b", hue='Food Category')

# Personalizar el gráfico
plt.xlabel("Año")
plt.ylabel("Desperdicio Total (Tons)")
plt.title("Evolución del Desperdicio Total en España (2014-2024)")
plt.grid(True, linestyle="--", alpha=0.7)



# Mostrar gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar datos solo para UK
data_UK = data_2[data_2['Country'] == "UK"]

# Configurar el gráfico
plt.figure(figsize=(10, 5))
sns.lineplot(data=data_UK, x="Year", y="Total Waste (Tons)", marker="o", color="b", hue='Food Category')

# Personalizar el gráfico
plt.xlabel("Año")
plt.ylabel("Desperdicio Total (Tons)")
plt.title("Evolución del Desperdicio Total en UK (2014-2024)")
plt.grid(True, linestyle="--", alpha=0.7)

# Mostrar gráfico
plt.show()
```
